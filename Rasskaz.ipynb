Сегодня тема про потребности в откровениях между вашими родными и близкими людьми
Откровенные разговоры в отношениях очень важная часть нашей жизни, и не важно про какие именно отношения идёт речь, будь то отношения родителей с детьми, пары девушка и парень или отношения мужа и жены, надо часто затевать откровенные разговоры, ведь когда этого не происходит то тут начинаются проблемы. И проблемы с далеким будущим и далеко не приятным откликом.
Возьмем к примеру родителей и детей. Ребенок в начале своего жизненного пути ещё не имеет опыта и порой не знает как поступить в определённых ситуациях, и что за собой влекут в дальнейшем определённые действия и поступки. И если у ребёнка не будет доверительных и откровенных разговоров с родителями и им из-за суеты будних дней и своих взрослых проблем будет всегда не до разговоров с ребёнком то ребёнок либо будет на своём горьком опыте набивать себе шишки и на подсознательном уровне запоминать что родитель ему не помощник и не защитник а это в будущем родителям ой как горько и больно откликнется!!! Или может этот ребёнок будет обращаться за опытом к посторонним людям, а тут не маловажную роль играет то кто будут эти люди какой уровень мышления или нравственного понимания у них. Хорошо если повезёт ребёнку что эти посторонние порядочные мудрые, а если это окажутся подворотние наркоманы или алкаши у которых полно свободного времени и вечно в своём одурманенном состоянии шатаются по дворам? Получается либо ваш ребёнок ведёт откровенные беседы со своими родителями либо с кем то другим, выбор за нами!!!
Теперь возьмём пример отношений между бизнес партнёрами либо между руководителем и подчинённым, тут тоже если у вас с подчинённым не будет налажен доверительный откровенный контакт или хотя бы вы относились с пониманием к сотруднику когда он(она) откровенно что то вам говорит касающееся рабочих моментов. То, сами понимаете, какие могут быть последствия…
Ну и на последок возьмём пример отношений в паре муж и жена. Тут тоже очень и очень важно разговаривать откровенно, да понимаю мы устаём на работе, нету настроения или сил, но надо хотя бы раз в неделю(а то и чаще) в какой нибудь выходной день взять себе за правило посвятить несколько часов разговоров с друг другом, откровенных разговоров, можете посидеть вдвоём при свечах и отключенных телефонов, поделиться планами или мыслями, переживаниями и прочими откровенными разговорами. И старайтесь с пониманием принять откровения вашего партнёра, и даже если это касается каких то интимных моментов, тут тоже отнеситесь с пониманием, да мы живём в век цифровой информации, да на просторах интернета что только мы может не увидеть и не всегда специально а часто даже случайно, и потом иногда хочется что то эдакое попробовать с партнёром ну или хотя бы поговорить об этом, да не стесняйтесь а проговаривайте даже интимные моменты это важно, ведь если вы будете отвергать и игнорировать такие темы то ваш партнёр(партнёрша) с кем то другим(другой) будет обсуждать а возможно и не только обсуждать… Либо вы закрываете или хотя-бы обсуждаете потребности вашего партнёра, либо(не всегда но часто) это сделает кто то на стороне, выбор за нами!!! 
И на мой взгляд порой не мешало бы с партнёром просто подурачиться в постели, устроить бой с подушками, пощекотать друг дружку, побаюкать, просто погладить по голове, по спине, и на мой взгляд очень важно обниматься.

Надо быть опорой, и поддержкой друг другу, попробуйте себя в роли семейного психолога можно даже в роли такой игры, слушайте внимательно своих близких когда они к вам обращаются с доверительным разговором!

Сегодня тема про наблюдательность путь к развитию, саморазвитию и самодисциплины
Если использовать этот навык по максимуму то можно достичь очень многого. 
Но с чего начинается работа над собой? А начинается всё с наблюдательности! Сначала мы просто наблюдаем за своими действиями, во сколько засыпаем и во сколько просыпаемся, вовремя ли питаемся, занимаемся ли физ упражнениями по утрам дома или вечером в фитнес зале, ставим ли хотя бы иногда план и строго ли мы ему следуем или относимся снисходительно. Наблюдаем ли мы за своими мыслями, позитивно ли мы мыслим или же нас часто сопровождают негативные, агрессивные или унылые мысли из-за которых наше моральное и эмоциональное состояние пребывает в упадке. Наблюдаем лы мы за своими поступками по отношению к окружающим нас людям, как часто мы справедливо и совестно поступаем по отношению к ним, наблюдаем лы мы за нашим отношениям с родителями, как они себя чувствуют, часто ли вы их навещаете, хорошо ли вы за ними заботитесь не смотря на ваши бытовые или иные разногласия? Сегодня тема про чистоту нашей речи!
Конечно же говорить можем много, разное, правду/кривду, ложь/истину, это оставим на совесть говорящего, сегодня наша тема слова паразиты.
Рассмотрим наиболее часто используемые популярные слова паразиты, начнём со слова "короче", которое люди очень очень часто используют пока с вами говорят и на самом деле не укорачивают а удлинняют свою речь, так что убрав из своего лексикона именноо это слово паразит ваша речь станет действительно короче )))
Следующее слово "в принципе", тоже не показвает нас с хорошей стороны, конечно же у людей есть принципы и это хорошо, ведь именно принцип порой нас удерживает на том что-бы не бросить начатое дело, либо дальше действовать согласно поставленному плану, но тут тоже надо знать меру ведь принцип пожно сравнить с якорем, иногда полезно стоять на якоре кораблю а иногда и вредно, ведь якорь держит корабль в одном месте и не даёт двигаться, так же и принципы иногда полезны удержать нас в чём-то а порой полезно отступать от своих принципов и менять свою точку мнения, по ходу роста наших знаний и навыков.
Теперь перейдём к следующему слову паразиту "если честно" или "честно говоря", то есть ты слушаешь своего собеседника, он(а) тебе рассказвает что-то на протяжении какого-то времени и тут вдруг прилетает слово "честно говоря" и далее следует новый поток слов. И если логически взглянуть на ситуацию то получается что весь разговор что был до этого слова можно множно смело выбробросить так как он был не честным, пустым и бесполезным, а вот сейчас начнется настоящий, истинный, честный разговор или рассказ? И получается этим словом паразитом мы обезцениваем весь наш предыдущий разговор или рассказ? Да и вообще выставляем себя лгуном который лишь иногда говорит честно?
Ну конечно же в большинстве случаев мы не являемся лгунами и упёртыми принципиалистами, и все эти многие другие слова паразиты мы слышим по радио и с экранов тв, особенно в озвучке на русский язык зарубежных фильмов. И надо работать над чистотой своей речи и избавляться от слов паразитов, и от матерных слов тем более!!! 
Сейчас прям какой-то тренд использовать в своей речи много слов состоящих из слов паразитов, а что самое ужасное матерных слов, а так же иностранных(типо модных) слов, и даже не задумываемся что те же англичане или американцы или испанцы в своей речи не используют ежедневно русские слова вводя их в свой модный тренд, нет нет, они не смешивают свои слова с русскими, а мы русскоговорящие люди с лёгкостью заменяем свою русскую речь иностранными словами, в какой то степени это предательство!!! 
Нам через моду навязывают идеологию пренебрежения всем русским и в то же самое время трепетное благоговение перед всем иностранным особенно европейским, у нас с вами незаметно затирают любовь ко всему русскому! Девушки в россии мечтают найти себе жениха европейца и парни из россии найти себе европейскую невесту и это не секрет, и жить в каком нибудь европейском государстве. Но подождите, а где любовь к своей стране, к своему роду к своим предкам к совоим традициям и обычиям, к своим девушкам и парням? Представим аналогию с домом, смотрим из окна своего дома на соседний и завидуем какой он у них красивый, и думаем вот бы жить там. Но, может мы станем любить свой дом?
