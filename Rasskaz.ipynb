Сегодня тема про потребности в откровениях между вашими родными и близкими людьми
Откровенные разговоры в отношениях очень важная часть нашей жизни, и не важно про какие именно отношения идёт речь, будь то отношения родителей с детьми, пары девушка и парень или отношения мужа и жены, надо часто затевать откровенные разговоры, ведь когда этого не происходит то тут начинаются проблемы. И проблемы с далеким будущим и далеко не приятным откликом.
Возьмем к примеру родителей и детей. Ребенок в начале своего жизненного пути ещё не имеет опыта и порой не знает как поступить в определённых ситуациях, и что за собой влекут в дальнейшем определённые действия и поступки. И если у ребёнка не будет доверительных и откровенных разговоров с родителями и им из-за суеты будних дней и своих взрослых проблем будет всегда не до разговоров с ребёнком то ребёнок либо будет на своём горьком опыте набивать себе шишки и на подсознательном уровне запоминать что родитель ему не помощник и не защитник а это в будущем родителям ой как горько и больно откликнется!!! Или может этот ребёнок будет обращаться за опытом к посторонним людям, а тут не маловажную роль играет то кто будут эти люди какой уровень мышления или нравственного понимания у них. Хорошо если повезёт ребёнку что эти посторонние порядочные мудрые, а если это окажутся подворотние наркоманы или алкаши у которых полно свободного времени и вечно в своём одурманенном состоянии шатаются по дворам? Получается либо ваш ребёнок ведёт откровенные беседы со своими родителями либо с кем то другим, выбор за нами!!!
Теперь возьмём пример отношений между бизнес партнёрами либо между руководителем и подчинённым, тут тоже если у вас с подчинённым не будет налажен доверительный откровенный контакт или хотя бы вы относились с пониманием к сотруднику когда он(она) откровенно что то вам говорит касающееся рабочих моментов. То, сами понимаете, какие могут быть последствия…
Ну и на последок возьмём пример отношений в паре муж и жена. Тут тоже очень и очень важно разговаривать откровенно, да понимаю мы устаём на работе, нету настроения или сил, но надо хотя бы раз в неделю(а то и чаще) в какой нибудь выходной день взять себе за правило посвятить несколько часов разговоров с друг другом, откровенных разговоров, можете посидеть вдвоём при свечах и отключенных телефонов, поделиться планами или мыслями, переживаниями и прочими откровенными разговорами. И старайтесь с пониманием принять откровения вашего партнёра, и даже если это касается каких то интимных моментов, тут тоже отнеситесь с пониманием, да мы живём в век цифровой информации, да на просторах интернета что только мы может не увидеть и не всегда специально а часто даже случайно, и потом иногда хочется что то эдакое попробовать с партнёром ну или хотя бы поговорить об этом, да не стесняйтесь а проговаривайте даже интимные моменты это важно, ведь если вы будете отвергать и игнорировать такие темы то ваш партнёр(партнёрша) с кем то другим(другой) будет обсуждать а возможно и не только обсуждать… Либо вы закрываете или хотя-бы обсуждаете потребности вашего партнёра, либо(не всегда но часто) это сделает кто то на стороне, выбор за нами!!! 
И на мой взгляд порой не мешало бы с партнёром просто подурачиться в постели, устроить бой с подушками, пощекотать друг дружку, побаюкать, просто погладить по голове, по спине, и на мой взгляд очень важно обниматься.
И тоже самое с ребёнком, подурачьтесь, поиграйте(смотря какой возраст ребёнка конечно же, ну вы что-нибудь придумайте) побудьте иногда с ним на его уровне не бойтесь спускаться со своего взрослого уровня до уровня(возраста) вашего ребёнка, и очень важно обнимать детей почаще и не важно на сколько он(она) уже взрослый. 
Будьте со своими близкими очень близко! Ваши дети должны делиться откровениями именно с вами а не с посторонними, ваш спутник жизни должен с вами делиться своими откровениями а не с кем то ещё! Не ругайте и не отчитывайте своих (и чужих тоже) детей когда они с вами делятся своими откровениями. Не критикуйте и не смущайте своего спутника жизни за его(её) откровения или пожелания, а разговаривайте и обсуждайте вместе. Надо быть опорой, и поддержкой друг другу, попробуйте себя в роли семейного психолога можно даже в роли такой игры, слушайте внимательно своих близких когда они к вам обращаются с доверительным разговором!
